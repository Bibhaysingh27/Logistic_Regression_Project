```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
df=pd.DataFrame(np.c_[cancer['data'],cancer['target']],columns=np.append(cancer['feature_names'],['target']))
```

```python
df.head()
```

```python
sns.pairplot(df,hue='target',vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])
plt.show()
```

```python
sns.countplot(df['target'])
plt.show()
```

```python
sns.scatterplot(x='mean area',y='mean smoothness',hue='target',data=df)
plt.show()
```

```python
df_cancer=df.drop(['target'],axis=1)
```

```python
df_cancer.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(df_cancer)
```

```python
scaled_features=pd.DataFrame(scalar.transform(df_cancer),columns=cancer['feature_names'])
```

```python
df_cancer.head()
```

```python
scaled_features.head()
```

```python
y=df['target']
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(scaled_features,y,test_size=0.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
y_test[:5]
```

```python
graph = model.predict_proba(X_test)[:, 0]
```

```python
for i in range(X_test.shape[1]):
    plt.scatter(X_test.iloc[:, i], y_test)
    plt.plot(X_test.iloc[:, i], pred, 'rx')
    plt.title(f"{X_test.columns[i]}")
    plt.plot(X_test.iloc[:, i], graph, 'c.')
    
    plt.show()
```
